```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/thyroid-disease/allbp.data')

df2=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/thyroid-disease/allhyper.data')
```

```python
df=pd.concat([df1,df2],axis=0)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.columns = ['age','sex','onthyroxine','querythyroxine','antimedi.','sick','pregnant','thyroid-surgery','I131','hypoth','hyperth',
                'lithium','goitre','tumor','hypoputit','psych','TSHm','TSH','T3m','T3','TT4m',
                    'TT4','T4Um','T4U','FTIm','FTI','TBGm','TBG','source','classes']
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.replace("?",np.NaN,inplace=True)
```

```python
df.isnull().sum()
```

```python
drop=df.isnull().sum()[df.isnull().sum()>len(df)*0.1]
df.drop(drop.index,axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
print(df['age'].mode())
print(df['TT4'].mode())
print(df['sex'].mode())
```

```python
df['age'].fillna('59',inplace=True)
df['TT4'].fillna('101',inplace=True)
df['sex'].fillna('F',inplace=True)
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=['sex','onthyroxine','querythyroxine','antimedi.','sick','pregnant','thyroid-surgery','I131','hypoth','hyperth',
                                'lithium','goitre','tumor','hypoputit','psych','TSHm','T3m','TT4m',
                                 'T4Um','FTIm','source'])
```

```python
df.columns
```

```python
df.drop(columns=['sex_M','onthyroxine_f','querythyroxine_f','antimedi._f','sick_f','pregnant_f','thyroid-surgery_f','I131_f','hypoth_f','hyperth_f','hypoputit_f','psych_f','TSHm_f','T3m_f','TT4m_f','T4Um_f','FTIm_f'],axis=1,inplace=True)
```

```python
df[['Class','Y']]=df.classes.str.split('.',expand=True,n=1)#It will split the column for first occurence of '.' and it will do it once as n=1,and save it in two different columns.
del df['Y']
del df['classes']
```

```python
feature={'Class': {'negative':1, 'increased binding protein':2,
       'decreased binding protein':3, 'hyperthyroid':4, 'T3 toxic':5, 'goitre':6}}
df.replace(feature)  # You can see the changes made.
```
